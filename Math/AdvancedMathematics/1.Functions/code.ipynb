```python
import math
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sgn(x):
    return 0 if x == 0 else -1 if x < 0 else 1
```

```python
x = np.linspace(-10, 10, 100)
y = [sgn(i) for i in x]
plt.plot(x, y)
```

```python
def round(x):
    return math.floor(x)
```

```python
x = np.linspace(-10, 10, 10000)
y = [round(i) for i in x]
plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 10000)
y = x
plt.plot(x, y)
plt.text(6, 9, r'$f\ (x) = x$')

x1 = np.linspace(0.01, 10, 10000)
y1 = np.log(x1)
plt.plot(x1, y1)
plt.text(7, 3, r'$f\ (x) = ln(x)$')

x2 = np.linspace(-10, 3, 10000)
y2 = np.exp(x2)
plt.plot(x2, y2)
plt.text(2, 5, r'$f\ (x) = e^{x}$')

plt.grid(True)
plt.axhline(color='black')
plt.axvline(color='black')
plt.show()
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
m = [-2, -1, 0, 1, 2]
x = np.linspace(-10, 10, 100)
for i in range(len(m)):
    ax = axes[i]
    ax.plot(x, [math.pow(n, m[i]) for n in x])
    ax.grid(True)
    ax.axhline(color='black')
    ax.axvline(color='black')
plt.show()
```

```python
fig, axes = plt.subplots(1, 5, figsize=(30, 4))
a = [2, 3, 4, 5, 6]
x = np.linspace(-10, 10, 100)
for i in range(len(a)):
    ax = axes[i]
    ax.plot(x, [math.pow(a[i], n) for n in x])
    ax.grid(True)
    ax.axhline(color='black')
    ax.axvline(color='black')
plt.show()
```

```python
fig, axes = plt.subplots(1, 5, figsize=(30, 4))
a = [2, 3, 4, 5, 6]
x = np.linspace(0.001, 10, 100)
for i in range(len(a)):
    ax = axes[i]
    ax.plot(x, [math.log(n, a[i]) for n in x])
    ax.grid(True)
    ax.axhline(color='black')
    ax.axvline(color='black')
plt.show()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 4))
x = np.linspace(-10, 10, 1000)
ax = axes[0]
ax.plot(x, np.sin(x))
ax.grid(True)
ax.axhline(color='black')
ax.axvline(color='black')
ax = axes[1]
ax.plot(x, np.cos(x))
ax.grid(True)
ax.axhline(color='black')
ax.axvline(color='black')
x = np.linspace(-5, 5, 100)
ax = axes[2]
ax.plot(x, np.tan(x))
ax.grid(True)
ax.axhline(color='black')
ax.axvline(color='black')
ax = axes[3]
ax.plot(x, np.cos(x) / np.sin(x))
ax.grid(True)
ax.axhline(color='black')
ax.axvline(color='black')
plt.show()
```
